Creating a Electricity Cost Prediction Model

```python
# Importing the required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# let x represent average daily usage (hrs) and y represent Monthly Bills ($)
x = np.array([[3],[4],[6],[7],[8],[10],[12]])
y = np.array([40, 54, 78, 97, 111, 135, 158])

# now to initialize the model: 
model = LinearRegression()
model.fit(x,y)

# Making Predections based on the model
predictions = model.predict(x)
print("Predicted Prices:", predictions)
```

```python
# Finding the intercept and Coefficient
print(f"Intercept: {model.intercept_}")
print(f"Coefficient/Slope: {model.coef_[0]}")
```

```python
# Predict bill 5 hours
predict_5 = model.predict(np.array([[5]]))
print("Bill for 5 Hrs: ",predict_5[0])

# Predict bill for 9 and 11 hoours 
predict_multi = model.predict(np.array([[9],[11],[15]]))
print("Bill for 9 Hrs: ",predict_multi[0])
print("Bill for 11 Hrs: ",predict_multi[1])
print("Bill for 15 Hrs: ",predict_multi[2])                          
```

```python
# Plot the data and regression line
# Predictions based on the model
predictions = model.predict(x)

# Plotting the data and the regression line
plt.figure(figsize=(8, 6))

# Scatter plot for actual data points
plt.scatter(x, y, color='blue', label='Actual Data')

# Plotting the regression line
plt.plot(x, predictions, color='red', label='Regression Line')
# Adding labels and title
plt.xlabel('Average Daily Usage (hours)')
plt.ylabel('Monthly Bill ($)')
plt.title('Electricity Bill Prediction using Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

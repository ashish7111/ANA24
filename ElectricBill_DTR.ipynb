```python
# Predecting spendings on electricity bill 

# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Data to be used for the model
# Average daily usage and monthly bill

x = ([[5],[7],[8],[11],[12],[14],[16]]) # Average daily hours
y = ([70, 98, 101, 138, 140, 161, 188 ])

# Initialize decision tree regressor
regressor = DecisionTreeRegressor(max_depth = 3, random_state = 42) 
# Limit for max_depth is set to 3 to reduce overfitting which leads to good results in the model but poor results in actual use
# random_state is to set a seed value so the same random choices are made each time
regressor.fit(x,y)

# Predicting the test data
x_test = np.linspace(5,16, 120).reshape(-1,1) 
''' Creating an array of 120 evenly spaced numbers between 5 and 16 inclusive based on our input and 
    converting it to a 2D array for use in Scikit-learn
'''
y_pred = regressor.predict(x_test)

# Predictions for specific daily usage values
new_data = np.array([[6], [9], [13]])  # New data points for prediction
new_predictions = regressor.predict(new_data)

# Print predictions
for i, hours in enumerate(new_data.flatten()):
    print(f"Predicted bill for {hours} hours/day: ${new_predictions[i]:.2f}")

#  Checking the accuracy of the model
y_train_pred = regressor.predict(x)
mse = mean_squared_error(y, y_train_pred)
print(f"Mean Squared Error = {mse}")

# Plotting the Decision Tree Regression
# Size of the figure for the plot
plt.figure(figsize=(8,6))
# Scatter plot for original Data
plt.scatter(x, y, color='blue', label = 'Actual Data', s =80)
# Regression line
plt.plot(x_test, y_pred, color='red', label ='Decision Tree Prediction')
plt.scatter(new_data, new_predictions, color='green', label='New Predictions', s=100)  # Predicted points
plt.xlabel('Average daily electricity Usage (Hrs)', fontsize = 12)
plt.ylabel('Electricity Bills ($)', fontsize = 12)
plt.title('Decision Tree Regression for Electricity Bill Prediction', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```
